```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager

# Load the CSV file
RF_df = pd.read_csv('Data/RF_Feature_Importance.csv')

# Rename columns
RF_df.columns = ['Feature', 'Importance']

# Calculate the total importance
total_importance = RF_df['Importance'].sum()

# Calculate the weight percentage
RF_df['Weight Percentage'] = (RF_df['Importance'] / total_importance) * 100

# Sort by weight percentage in descending order
df_sorted = RF_df.sort_values(by='Weight Percentage', ascending=False)

# Plotting the bar chart with updated styles
plt.figure(figsize=(18, 12))
plt.barh(df_sorted['Feature'], df_sorted['Weight Percentage'], color='Thistle', linewidth=0.5, height=0.73)  # Adjust the height parameter

# Set custom fonts
font_prop = font_manager.FontProperties(size=16)

plt.xlabel('Weight Percentage (%)', fontsize=18, labelpad=20, fontproperties=font_prop)
plt.ylabel('Feature', fontsize=20, labelpad=20, fontproperties=font_prop)
plt.title('Random Forest Feature Importance Ranking', fontsize=19, pad=20)

# Add a grey dashed grid
plt.grid(axis='x', color='gray', linestyle='--', linewidth=0.5)

# Customize ticks
plt.xticks(fontsize=16)
plt.yticks(fontsize=15)

# Adjust bar spacing
plt.subplots_adjust(left=0.25, right=0.9, top=0.9, bottom=0.1)

# Invert y-axis to have the highest importance on top
plt.gca().invert_yaxis()

# Save the figure to a file in the same directory as the notebook
output_path = 'Random Forest Feature Importance Ranking.png'
plt.savefig(output_path, bbox_inches='tight') 
plt.show()

# Display the sorted dataframe (optional)
print(df_sorted[['Feature', 'Weight Percentage']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager

# Load the CSV file
GBT_df = pd.read_csv('Data/GBT_Feature_Importance.csv')

# Rename columns
GBT_df.columns = ['Feature', 'Importance']

# Remove commas from the 'Importance' column and convert to numeric
GBT_df['Importance'] = GBT_df['Importance'].str.replace(',', '').astype(float)

# Calculate the total importance
total_importance = GBT_df['Importance'].sum()

# Calculate the weight percentage
GBT_df['Weight Percentage'] = (GBT_df['Importance'] / total_importance) * 100

# Sort by weight percentage in descending order
df_sorted = GBT_df.sort_values(by='Weight Percentage', ascending=False)

# Plotting the bar chart with updated styles
plt.figure(figsize=(18, 12))
plt.barh(df_sorted['Feature'], df_sorted['Weight Percentage'], color='LightGreen', linewidth=0.5, height=0.73)  # Adjust the height parameter

# Set custom fonts
font_prop = font_manager.FontProperties(size=16)

plt.xlabel('Weight Percentage (%)', fontsize=18, labelpad=20, fontproperties=font_prop)
plt.ylabel('Feature', fontsize=20, labelpad=20, fontproperties=font_prop)
plt.title('Gradient Boosting Tree Feature Importance Ranking', fontsize=19, pad=20)

# Add a grey dashed grid
plt.grid(axis='x', color='gray', linestyle='--', linewidth=0.5)

# Customize ticks
plt.xticks(fontsize=16)
plt.yticks(fontsize=15)

# Adjust bar spacing
plt.subplots_adjust(left=0.25, right=0.9, top=0.9, bottom=0.1)

# Invert y-axis to have the highest importance on top
plt.gca().invert_yaxis()

# Save the figure to a file in the same directory as the notebook
output_path = 'Gradient Boosting Tree Feature Importance Ranking.png'
plt.savefig(output_path, bbox_inches='tight') 
plt.show()

# Display the sorted dataframe (optional)
print(df_sorted[['Feature', 'Weight Percentage']])
```
